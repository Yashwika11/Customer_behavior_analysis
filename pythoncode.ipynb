```python
import pandas as pd
```

```python
df = pd.read_csv('customer_shopping_behavior.csv')
```

```python
df.isnull().sum()
```

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
#feature engineering
# creating a column age_group
labels = ['Young','Adult','Middle-aged','Senior']
df['age_group'] = pd.qcut(df['age'], q=4 , labels = labels)
```

```python
df[['age','age_group']].head(12)
```

```python
#creating column purchase_frequency_days
frequency_mapping = {
    'Daily': 1 ,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly':90 ,
    'Bi-Weekly':14 ,
    'Anually' : 365 
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping).astype('Int64')
```

```python
from sqlalchemy import create_engine
from dotenv import load_dotenv
import os

load_dotenv(".env", override=True)

engine = create_engine(
    f"mysql+pymysql://{os.getenv('DB_USERNAME')}:{os.getenv('DB_PASSWORD')}@{os.getenv('DB_HOST')}:{os.getenv('DB_PORT')}/{os.getenv('DB_NAME')}"
)

with engine.connect() as conn:
    print("Connection successful!")

df.to_sql("customer", engine, if_exists="replace", index=False)

print("Data loaded successfully!")
```
